```python
import numpy as np
import pandas as pd 
import os 
import scanpy as sc 
import matplotlib.pyplot as plt
```

```python
import h5py

allen_23 = pd.read_csv('allen_23_var.csv')

f = h5py.File(r"D:\sly_data\mus_brain_ref_data\CTX_Hip_counts_10x.h5",'r')
allen_21 = f['data']['gene'][:].astype('U') 

mus_pri = pd.read_csv('features.tsv',sep = '\t', header= None)
gex_pri = mus_pri.loc[mus_pri.iloc[:,2] == 'Gene Expression',:] 

assert (gex_pri.iloc[:,0] == allen_23.iloc[:,0]).all()
assert (gex_pri.iloc[:,1] == allen_23.iloc[:,1]).all()
```

```python
gex_pri.index = gex_pri.loc[:,0].values
gex_pri.nunique()
```

```python
# make gene_symbol unique  

gene_symbol = gex_pri[1].value_counts().index.values

duplicated = gene_symbol[gex_pri[1].value_counts() > 1]
print(duplicated.shape)
print(duplicated[:10])

duplicated_idx = gex_pri.index[gex_pri[1].isin(duplicated)]

for val in duplicated:
    sub_df = gex_pri[gex_pri[1] == val]
    for i in range(sub_df.shape[0]):
        if i == 0: 
            continue 
        else:
            idx = sub_df.index[i]
            gex_pri.loc[idx,1] = gex_pri.loc[idx,1] + f'_{i}'
gex_pri.nunique()
```

```python
mutual_gene = np.intersect1d(gex_pri[1], allen_21)
print(mutual_gene.shape)
```

```python
gene_dic = {}
for i,gene in enumerate(mutual_gene):
    gene_dic[gene] = i
gene_dic
```
